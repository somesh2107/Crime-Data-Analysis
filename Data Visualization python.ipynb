```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
file_path = 'Homework_1_Data_Mining.csv'
df = pd.read_csv(file_path)

# Assuming 'CrimeDate' is in a format like 'YYYY-MM-DD'
df['CrimeDate'] = pd.to_datetime(df['CrimeDate'])
df['Year'] = df['CrimeDate'].dt.year
df['Month'] = df['CrimeDate'].dt.month
df['DayOfWeek'] = df['CrimeDate'].dt.dayofweek
```

```python
# Yearly crimes
df['Year'].value_counts().sort_index().plot(kind='bar', title='Year-to-Year Crime Incidences')
plt.xlabel('Year')
plt.ylabel('Number of Crimes')
plt.show()

# Monthly crimes
df['Month'].value_counts().sort_index().plot(kind='bar', title='Monthly Crime Incidences')
plt.xlabel('Month')
plt.ylabel('Number of Crimes')
plt.show()

# Day of Week crimes
df['DayOfWeek'].value_counts().sort_index().plot(kind='bar', title='Day-of-Week Crime Incidences')
plt.xlabel('Day of Week')
plt.ylabel('Number of Crimes')
plt.xticks(ticks=range(7), labels=['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun'])
plt.show()
```

```python
# Distribution of crime types
df['Description'].value_counts().plot(kind='pie', title='Distribution of Crime Types', autopct='%1.1f%%')
plt.ylabel('')  # Remove the 'Description' label on the y-axis for clarity
plt.show()
```

```python
# Yearly distribution of crime types
for year in df['Year'].unique():
    df_year = df[df['Year'] == year]
    df_year['Description'].value_counts().plot(kind='bar', title=f'Crime Type Distribution in {year}')
    plt.xlabel('Crime Type')
    plt.ylabel('Number of Crimes')
    plt.show()

# This pattern can be repeated for months and days of the week
```

```python
import matplotlib.pyplot as plt

# Month names for plotting
months = {1: 'January', 2: 'February', 3: 'March', 4: 'April', 5: 'May', 6: 'June',
          7: 'July', 8: 'August', 9: 'September', 10: 'October', 11: 'November', 12: 'December'}

# Plotting crime type distribution for each month
for month in sorted(df['Month'].unique()):
    df_month = df[df['Month'] == month]
    df_month['Description'].value_counts().plot(kind='bar', title=f'Crime Type Distribution in {months[month]}')
    plt.xlabel('Crime Type')
    plt.ylabel('Number of Crimes')
    plt.show()
```

```python
# Day names for plotting
days_of_week = {0: 'Monday', 1: 'Tuesday', 2: 'Wednesday', 3: 'Thursday', 
                4: 'Friday', 5: 'Saturday', 6: 'Sunday'}

# Plotting crime type distribution for each day of the week
for day in sorted(df['DayOfWeek'].unique()):
    df_day = df[df['DayOfWeek'] == day]
    df_day['Description'].value_counts().plot(kind='bar', title=f'Crime Type Distribution on {days_of_week[day]}')
    plt.xlabel('Crime Type')
    plt.ylabel('Number of Crimes')
    plt.show()
```
