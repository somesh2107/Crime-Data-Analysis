```python
import pandas as pd

# Load the uploaded CSV file
file_path = 'Homework 1 Data Mining.csv'
df = pd.read_csv(file_path)

# Calculate the necessary statistics
min_values = df.select_dtypes(include=[float, int]).min()
max_values = df.select_dtypes(include=[float, int]).max()
mean_values = df.select_dtypes(include=[float, int]).mean()
median_values = df.select_dtypes(include=[float, int]).median()
std_values = df.select_dtypes(include=[float, int]).std()
total_rows = df.shape[0]
missing_values = df.isnull().sum()
percent_missing = (missing_values / total_rows) * 100

# Create the DataFrame with the desired information
df_info = pd.DataFrame({
    'Data Type': df.dtypes,
    'Missing Values': missing_values,
    'Percent Missing': percent_missing,
    'Unique Values': df.nunique(),
    'Min Value': min_values.reindex(df.columns),
    'Max Value': max_values.reindex(df.columns),
    'Mean': mean_values.reindex(df.columns),
    'Median': median_values.reindex(df.columns),
    'Standard Deviation': std_values.reindex(df.columns)
})

df_info
```

## Here you cana see Column Location1, vri_name1 and Weapon has nearly about more than 80% of the values missing so we can drop that tables as it will not be that important in the analysis 

## Deleting the columns having missing values more than 80%

```python
# Drop the columns 'Premise' and 'Weapon'
df = df.drop(columns=['Weapon', 'Location 1', 'vri_name1'])
```

```python
df
```

## Deleting the missing values where missing values are less than 5 percent 

```python
# Calculate the threshold for maximum allowed missing values
threshold = 0.05 * len(df)

# Identify columns where missing values are less than 5% of the dataset
cols_to_consider = df.columns[df.isnull().sum() < threshold]

# Drop rows where any of the identified columns have missing values
df_cleaned = df.dropna(subset=cols_to_consider)

# Save or proceed with the cleaned DataFrame
df_cleaned.head()  # Display the first few rows of the cleaned DataFrame
```

```python
# Calculate the necessary statistics
min_values = df_cleaned.select_dtypes(include=[float, int]).min()
max_values = df_cleaned.select_dtypes(include=[float, int]).max()
mean_values = df_cleaned.select_dtypes(include=[float, int]).mean()
median_values = df_cleaned.select_dtypes(include=[float, int]).median()
std_values = df_cleaned.select_dtypes(include=[float, int]).std()
total_rows = df_cleaned.shape[0]
missing_values = df_cleaned.isnull().sum()
percent_missing = (missing_values / total_rows) * 100

# Create the DataFrame with the desired information
df_cleaned_info = pd.DataFrame({
    'Data Type': df_cleaned.dtypes,
    'Missing Values': missing_values,
    'Percent Missing': percent_missing,
    'Unique Values': df_cleaned.nunique(),
    'Min Value': min_values.reindex(df_cleaned.columns),
    'Max Value': max_values.reindex(df_cleaned.columns),
    'Mean': mean_values.reindex(df_cleaned.columns),
    'Median': median_values.reindex(df_cleaned.columns),
    'Standard Deviation': std_values.reindex(df_cleaned.columns)
})

df_cleaned_info
```

```python
# Identify the mode of the 'Inside/Outside' and 'Premise' columns
inside_outside_mode = df_cleaned['Inside/Outside'].mode()[0]
premise_mode = df_cleaned['Premise'].mode()[0]

print(inside_outside_mode)
print(premise_mode)

df_cleaned['Inside/Outside'] = df_cleaned['Inside/Outside'].fillna(inside_outside_mode)
df_cleaned['Premise'] = df_cleaned['Premise'].fillna(premise_mode)

# Calculate the necessary statistics
min_values = df_cleaned.select_dtypes(include=[float, int]).min()
max_values = df_cleaned.select_dtypes(include=[float, int]).max()
mean_values = df_cleaned.select_dtypes(include=[float, int]).mean()
median_values = df_cleaned.select_dtypes(include=[float, int]).median()
std_values = df_cleaned.select_dtypes(include=[float, int]).std()
total_rows = df_cleaned.shape[0]
missing_values = df_cleaned.isnull().sum()
percent_missing = (missing_values / total_rows) * 100

# Create the DataFrame with the desired information
df_cleaned_info = pd.DataFrame({
    'Data Type': df_cleaned.dtypes,
    'Missing Values': missing_values,
    'Percent Missing': percent_missing,
    'Unique Values': df_cleaned.nunique(),
    'Min Value': min_values.reindex(df_cleaned.columns),
    'Max Value': max_values.reindex(df_cleaned.columns),
    'Mean': mean_values.reindex(df_cleaned.columns),
    'Median': median_values.reindex(df_cleaned.columns),
    'Standard Deviation': std_values.reindex(df_cleaned.columns)
})

df_cleaned_info
```

## Now you can see the data is cleaned it has no missing values we can now start analysis on it in tableau

```python
# Export the cleaned DataFrame to a new CSV file
output_file_path = 'Cleaned_Homework_1_Data_Mining.csv'
df_cleaned.to_csv(output_file_path, index=False)

print(f"Data exported successfully to {output_file_path}")
```
